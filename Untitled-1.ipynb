```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Time variable (e.g., 100 time points)
t = np.linspace(0, 10, 1000)

# Simulated CPU usage for 4 servers with slight variations
# Values fluctuate between 0 and 100
cpu1 = 50 + 20 * np.sin(t) + 5 * np.random.randn(1000)  # Base at 50% with slight fluctuation
cpu2 = 55 + 8 * np.sin(t + 0.5) + 4 * np.random.randn(1000)  # Slightly shifted and different amplitude
cpu3 = 45 + 12 * np.sin(t + 1.0) + 6 * np.random.randn(1000)  # Slightly different pattern
cpu4 = 60 + 30 * np.sin(t + 1.5) + 3 * np.random.randn(1000)  # Different base and phase shift

# Ensure the CPU usage stays within the 0-100 range by clipping
cpu1 = np.clip(cpu1, 0, 100)
cpu2 = np.clip(cpu2, 0, 100)
cpu3 = np.clip(cpu3, 0, 100)
cpu4 = np.clip(cpu4, 0, 100)

# Create a figure and axis for each subplot
fig, axes = plt.subplots(4, 1, figsize=(4, 3), sharex=True)

# Plot each time series on a separate axis
axes[0].plot(t, cpu1, color='b')
axes[1].plot(t, cpu2, color='r')
axes[2].plot(t, cpu3, color='g')
axes[3].plot(t, cpu4, color='purple')

# Remove x and y-axis numbers (ticks and labels) and spines for each subplot
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

    # Turn off the spines (rectangles around the graph)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)


plt.tight_layout()

plt.savefig('cpu_cluster_stacked.png')

# Display the plot
plt.show()
```
